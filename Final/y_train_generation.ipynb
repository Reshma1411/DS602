```python
import pandas as pd

# Define your SME class with the get() and ask() methods
class SME:
    def __init__(self):
        self.asked = 0
        self.df = self.get_data()  # Load your data (X_train and y_train) here

    def get_data(self):
        # Load your X_train and y_train data from the provided URLs
        X_train_url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/final3/X_train.csv'
        y_train_url = 'https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/final3/y_train.csv'
        
        X_train = pd.read_csv(X_train_url)
        y_train = pd.read_csv(y_train_url)
        
        # Merge X_train and y_train on some common identifier (e.g., index or ID column)
        X_train['Left'] = y_train['Left']  # Assuming 'Left' is your label column
        return X_train

    def ask(self, gender, distance, years_worked, hire_date, prev_salary, current_salary, dep_code):
        # Query your SME for labels based on input features
        # Implement your querying logic here using self.df
        # For example:
        query_result = self.df.query(f"Gender == {gender} and Distance == '{distance}' and YearsWorked == {years_worked} and HireDate == '{hire_date}' and PreviousSalary == '{prev_salary}' and Salary == '{current_salary}' and DepartmentCode == {dep_code}")
        
        if not query_result.empty:
            # Return the label from the query result (assuming 'Left' column is your label)
            return query_result.iloc[0]['Left']
        else:
            return None  # Handle case where no match is found

# Instantiate your SME class
sme = SME()

# Assuming X_train_sample is your DataFrame containing 500 randomly selected rows
# Load or create X_train_sample DataFrame here

# Iterate over each row in X_train_sample to query SME for labels
table_labels = []
for index, row in X_train_sample.iterrows():
    try:
        # Extract features from the row
        gender = row['Gender']
        distance = row['Distance']
        years_worked = row['YearsWorked']
        hire_date = row['HireDate']
        prev_salary = row['PreviousSalary']
        current_salary = row['Salary']
        dep_code = row['DepartmentCode']
        
        # Query SME for label
        label = sme.ask(gender, distance, years_worked, hire_date, prev_salary, current_salary, dep_code)
        
        # Append the row with the predicted label to table_labels
        row_with_label = row.copy()
        row_with_label['Label'] = label
        table_labels.append(row_with_label)
    except Exception as e:
        print(f"Error querying SME for row {index}: {e}")

# Create a DataFrame from table_labels
table_labels_df = pd.DataFrame(table_labels)

# Now table_labels_df contains X_train_sample with predicted labels from SME
print(table_labels_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Load the DataFrame with labels
# Assuming table_labels_df contains the DataFrame with features and labels

def convert_salary(df, columns):
    for col in columns:
        df[col] = df[col].apply(lambda x: float(x.replace('K', '')) * 1000 if 'K' in x else float(x))
    return df

columns_to_convert = ['PreviousSalary', 'Salary']
table_labels_df = convert_salary(table_labels_df, columns_to_convert)

# Define categorical and numerical features (assuming 'Distance' and 'DepartmentCode' are categorical)
categorical_features = ['Distance', 'DepartmentCode','Gender']
numerical_features = ['YearsWorked', 'PreviousSalary', 'Salary']

table_labels_df
```

```python
categorical_features = ['Distance', 'DepartmentCode','Gender']
numerical_features = ['YearsWorked', 'PreviousSalary', 'Salary','Label']
```

```python
numerical_imputer = SimpleImputer(strategy='median')
table_labels_df[numerical_features] = numerical_imputer.fit_transform(table_labels_df[numerical_features])
categorical_imputer = SimpleImputer(strategy='most_frequent')
table_labels_df[categorical_features] = categorical_imputer.fit_transform(table_labels_df[categorical_features])

table_labels_df_encoded = pd.get_dummies(table_labels_df, columns=categorical_features)
# Separate features (X) and target variable (y)
X = table_labels_df_encoded.drop(columns=['Label','HireDate'])  # Features
y = table_labels_df_encoded['Label']  # Target variable

```

```python

# Choose a classification model
model = RandomForestClassifier()  # Change to the desired model (e.g., LogisticRegression, SVC)

# Create a pipeline with preprocessing and the chosen model
pipeline = Pipeline([
    ('classifier', model)
])

# Perform cross-validation (5-fold cross-validation) to evaluate the model
cv_scores = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy')

# Print the accuracies obtained from cross-validation
print("Cross-Validation Accuracies:")
for fold_idx, accuracy in enumerate(cv_scores, start=1):
    print(f"Fold {fold_idx}: {accuracy:.4f}")

# Calculate and print the mean accuracy and standard deviation of cross-validation scores
mean_accuracy = cv_scores.mean()
std_accuracy = cv_scores.std()
print(f"\nMean Accuracy: {mean_accuracy:.4f} (+/- {std_accuracy:.4f})")
```

```python
# Define models to evaluate
models = [
    ('Random Forest', RandomForestClassifier()),
    ('Logistic Regression', LogisticRegression()),
    ('Support Vector Machine', SVC())
]

# Evaluate each model using cross-validation
results = []
for name, model in models:
    pipeline = Pipeline([
        ('classifier', model)
    ])
    
    # Perform cross-validation (5-fold cross-validation)
    cv_scores = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy')
    
    # Calculate mean and standard deviation of cross-validation scores
    mean_accuracy = cv_scores.mean()
    std_accuracy = cv_scores.std()
    
    # Store results
    results.append({
        'Model': name,
        'Mean Accuracy': mean_accuracy,
        'Std Deviation': std_accuracy
    })

# Display results and identify the best model based on mean accuracy
results_df = pd.DataFrame(results)
print("Model Evaluation Results:")
print(results_df)

# Identify the best model based on mean accuracy
best_model = results_df.loc[results_df['Mean Accuracy'].idxmax()]
print("\nBest Model:")
print(best_model)
```
