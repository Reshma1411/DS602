# Week11 - Clustering Lab

* Create clusters for the following dataset
* Choose a meaningful cluster count
* Identify potential labels for each cluster

```python
from zipfile import ZipFile
from io import BytesIO
import urllib.request as urllib2
import pandas as pd

def get_data():
    r = urllib2.urlopen("https://github.com/msaricaumbc/DS_data/blob/master/ds602/people_wiki.zip?raw=true").read()
    file = ZipFile(BytesIO(r))
    people_wiki_csv = file.open("people_wiki.csv")
    people = pd.read_csv(people_wiki_csv)
    people.drop('URI', inplace=True, axis=1)
    return people

people = get_data()
people.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(people['text'])
```

```python
# Choose a meaningful cluster count
num_clusters = 5
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=num_clusters, random_state=123)
kmeans.fit(X)
```

```python
# Get the cluster labels for each data point
cluster_labels = kmeans.labels_
```

```python
from collections import Counter
cluster_potential_labels = {}
```

```python
for cluster_id in range(num_clusters):
    # Get the data points belonging to the current cluster
    cluster_data = people[cluster_labels == cluster_id]
    
    word_counts = cluster_data['text'].apply(lambda x: Counter(x.split()))
    common_words = word_counts.sum().most_common(5)  # Get the top 5 most common words
    
    # Store the potential labels for the cluster
    cluster_potential_labels[cluster_id] = [word for word, count in common_words]
```

```python
# Print the potential labels for each cluster
for cluster_id, labels in cluster_potential_labels.items():
    print(f"Cluster {cluster_id}: {', '.join(labels)}")
```
